```python
# Ejercicio 1 Importamos las librerías que vamos a tener que utilizar
import mysql.connector
import pandas as pd
```

```python
# Ejercicio 2: Creamos Schema, manejamos la excepción porque como ya lo habremos creado la primera vez que ejecutemos la celda no podemos volver a crearlo

cnx = mysql.connector.connect(user='root', password='AlumnaAdalab', host='127.0.0.1')
cnx.is_connected()

mycursor = cnx.cursor()
try:
    mycursor.execute("CREATE SCHEMA Alumnas_Promo_X")
    
except :
    print("No se puede crear porque ya has creado el Schema")

print(mycursor)
```

```python
# Ejercicio 3: Creamos Tabla alumnas, manejamos la excepción porque como ya la habremos creado la primera vez que ejecutemos la celda no podemos volver a crearla

cnx = mysql.connector.connect(user='root', password='AlumnaAdalab', host='127.0.0.1', database = 'Alumnas_Promo_X')
mycursor = cnx.cursor()
query = ("CREATE TABLE alumnas (id_alumnas INT, nombre VARCHAR(20), apellidos VARCHAR (40), ciudad VARCHAR (20), teléfono INT)")

try:
    mycursor.execute(query)
except:
    print("No se puede volver a crear una tabla que ya has creado")

print(mycursor)


```

```python
# Ejercicio 4: Añadimos elementos a la tabla, si ejecutamos esta celda n veces, añadiremos todos estos valores cada vez que la ejecutemos

sql = "INSERT INTO alumnas (id_alumnas, nombre, apellidos, ciudad, teléfono) VALUES (%s,%s,%s,%s,%s)"
val = [
    (1, "Maria", "Valverde", "Barcelona", 123456789),
    (2, "Ana", "Romero", "Valencia", 234567891),
    (3, "Cesar", "Palacio", "Quito", 345678901),
    (4, "Pepe", "Viñas", "Salamanca", 456789012),
    (5, "Ramon", "Garcia", "Málaga", 567890123),
    (6, "Raúl", "Perez","Girona", 678901234), 
    (7, "Esther", "Tapiador", "Madrid", 789012345),
    (8, "Elena", "Simón", "Oviedo", 890123456),
    (9, "Silvia", "Garcia", "Logroño", 901234567),
    (10, "Rodrigo", "Sanz", "Lima", 987654321)
    ]

mycursor.executemany(sql,val)
cnx.commit()
print(mycursor)
```

```python
# Ejercicio 5: Primero usamos el metodo fetchall para poder acceder a los datos y guardarlos en una lista de tuplas, que son las que luego podremos 
# convertir en una tabla usando pandas. 

cnx = mysql.connector.connect(user = "root", password = "AlumnaAdalab", host ="127.0.0.1", database = "Alumnas_Promo_X")
mycursor = cnx.cursor()
query = ("SELECT * FROM alumnas")
mycursor.execute(query)

results = mycursor.fetchall()
print(results)
```

```python
# Una vez que tenemos la lista con las tuplas que contienen los registros de cada fila, lo pasamos como argumento al metodo Dataframe
# y especificamos las columnas a las que pertenecen estos registros. 


df = pd.DataFrame(results, columns=["id_alumnas","nombre", "apellidos", "ciudad", "teléfono"])
df
```

```python
# Ejercicio 7 (al preguntar si hay duplicados deberia retornarnos True, porque he insertado los mismos registros varias veces)

print(df.duplicated()) 

# Los primeros 10 elementos me dice que no están duplicados, pero a continuacion he metido estos elementos 2 veces más, así
# que los 20 restantes me dice que si están duplicados.
```

```python
# Ejercicio 9
# hacemos una nueva consulta para buscar que ciudades tenemos en la tabla, uso distinct para que no me muestra ciudades repetidas
cnx = mysql.connector.connect(user = "root", password = "AlumnaAdalab", host ="127.0.0.1", database = "Alumnas_Promo_X")
mycursor = cnx.cursor()
query = ("SELECT DISTINCT ciudad FROM alumnas")
mycursor.execute(query)

ciudades = mycursor.fetchall()
print(ciudades)
```

```python
# Ejercicio 10 

cnx = mysql.connector.connect(user = "root", password = "AlumnaAdalab", host ="127.0.0.1", database = "Alumnas_Promo_X")
mycursor = cnx.cursor()

query_1 = ("SELECT * FROM alumnas WHERE ciudad = 'Barcelona'")
query_2 = ("SELECT * FROM alumnas WHERE ciudad = 'Málaga'")

mycursor.execute(query_1)

ciudad_1 = mycursor.fetchall()

mycursor.execute(query_2)
ciudad_2 = mycursor.fetchall()

df1 = pd.DataFrame(ciudad_1)
df2 = pd.DataFrame(ciudad_2)

print(df1)
print(df2)
```
