```python
# Ejercicio 1 => Cread tres variables numéricas de tipo integer (entero), llamadas de la siguiente forma: numero1, numero2, numero3

numero1 = 7
numero2 = 75
numero3 = 14
```

```python
# Ejercicio 2 => Utilizando los operadores lógicos:

'''Numero 1 mayor que numero 2'''
print(numero1 > numero2) 

'''Es el numero 1 menor que el numero2 pero mayor que el numero3'''
print(numero3 < numero1 < numero2)
# Otra forma => 
print((numero1 < numero2) and (numero1 > numero3))

'''Es el numero3 igual que el numero2'''
print(numero3 == numero2)

'''Es el numero2 distinto que el numero1'''
print(numero2 != numero1)
```

```python
# Ejercicio 3: Cread dos variables numéricas de tipo float (decimal) con dos decimales, llamdas de la siguiente forma (altura1 y altura2):

altura1 = 205

altura2 = 163
```

```python
# Ejercicio 4 => Utilizando los operadores matemáticos aprendidos en la lección:

'''Cual es la suma de las 2 alturas'''
suma = altura1 + altura2
print(suma)

'''Cuál es la diferencia entre las alturas'''
diferencia = altura1 - altura2
print(diferencia)

'''Cuál es el resto de la división de las dos alturas'''
resto = altura1 % altura2
print(resto)

'''Cuál es resultado de la división de las dos alturas, redondead el resultado a un decimal.'''
division = round(altura1/altura2,1)
print(division)
```

```python
# Ejercicio 4: Escribid un programa usando la función input() que le pregunte al usuario su nombre y apellidos. 
# Una vez que tengamos el nombre del usuario:

nombre = input("Dime tu nombre y apellidos=>: ")

''' Printead su nombre y apellidos todo en mayúsculas '''
print(nombre.upper())

''' Printead su nombre y apellidos todo en minúsculas''' 
print(nombre.lower())

''' Printead su nombre y apellidos con la primera letra del nombre y los apellidos en mayúscula y el resto en minúscula.'''
print(nombre.title())

''' Printead solo la primera letra del nombre en mayúscula.'''
print(nombre.capitalize())
```

```python
# Ejercicio 5: Escribid un programa usando la función input() que:

''' Le pregunte al usuario por una frase, almacenad este resultado en una variable.'''
frase = input("Dime una frase => ")

''' Le pregunte al usuario por una vocal, almacenad este resultado en una variable.'''
vocal = input("Dame una vocal => ")

''' Printead la misma frase pero con la vocal introducida en mayúsculas.'''
print(frase.lower().replace(vocal, vocal.upper()))
```

```python
# Ejercicio 6: Escribid un programa usando la función input() que:

''' Le pregunte al usuario por la lista de la compra, separando cada elemento por comas'''
lista_compra = input("Dame una lista de la compra, separa cada ingrediente por comas. => ") 
lista_compra_separada = lista_compra.split(",")
print(lista_compra_separada)

'''Devolved la lista de compra donde veamos cada elemento en una línea.'''
print("\n".join(lista_compra_separada))
```
