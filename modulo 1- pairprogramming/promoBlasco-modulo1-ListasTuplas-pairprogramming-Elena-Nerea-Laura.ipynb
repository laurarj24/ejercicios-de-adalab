```python
# Ejercicio 13: Obtener una lista inversa

''' Descripción: Dada una lista, obtén una nueva lista que contenga los elementos en orden inverso.'''

lista3 = [0, 1, 1, 2, 3, 5, 8, 13]
lista3.reverse()
print(lista3)
```

```python
# Ejercicio 19: Agregar elementos a una lista utilizando append y extend

''' Descripción: Dada una lista inicial, agrega nuevos elementos utilizando tanto append como extend 
y observa las diferencias en el resultado final. El elemento que deberéis usar para añadir a la 
lista original tiene que ser una lista.'''

lista = [1,2,3]
lista1 = [11,11,11]
lista.append(lista1)
print(lista)

lista2 = [12,13,14]

lista2.extend(lista1)
print(lista2)
```

```python
# Ejericio 20: Insertar un elemento en una posición específica

''' Descripción: Dada una lista, inserta un nuevo elemento en una posición específica determinada 
por el usuario (debereis usar la función input para preguntar al usuario en que posición quiere 
insertar el elemento).'''

lista3 = ["tomates", "peras", "limones"]
posicion = int(input("Dime en qué posición quieres que añadamos a la lista de la compra el queso")) 
# lo convierto a int porque el input devuelve un string

lista3.insert(posicion, "queso")
print(lista3)

# Si el usuario da una posición superior al numero de indices de nuestra lista, lo que hace es añadirlo al final. 
```

```python
# Ejercicio 21: Verificar si un elemento está presente en una lista

''' Descripción: Dada una lista, verifica si un elemento específico está presente en ella. 
Igual que el caso anterior, debereis preguntar al usuario que elemento quiere buscar en la lista. 
El resultado deberá ser True o False.'''

lista3 = ['tomates', 'peras', 'limones', 'queso']
elemento_en_lista3 = input("Dime algun producto que quieras comprar para verificar si está en la lista de la compra").lower()
print(elemento_en_lista3 in lista3)
```

```python
# Ejercicio 22: Manipulación de una lista utilizando remove(), pop() y clear()

''' Descripción: Crea una lista inicial, realiza operaciones de eliminación y 
limpieza utilizando los métodos remove(), pop() y clear(), y observa cómo afectan a la lista.'''

personas = ["Laura", "Maria", "Fatima", "Irene", "Eugenia"]
print(personas)

personas.remove("Laura") # Elimina el elemento cuyo valor coincide con lo que nosotros le demos
print(personas)

personas.pop(3) # Elimina el elemento que está en la posición que le indicamos
print(personas)

personas.clear() # Vacía el contenido de la lista
print(personas)
```

```python
# Ejercicio 4: Concatenar dos tuplas

''' Descripción: Dadas dos tuplas, crea una nueva tupla que contenga todos los elementos de ambas tuplas.'''

tupla_1 = (2, 7, "Maria", True, 48, "Irene")
tupla_2 = ("años", "meses", False, 35)

tupla_mix = tupla_1 + tupla_2
print(tupla_mix)
```

```python
# Ejercicio 5: Encontrar el índice de un elemento en una tupla

''' Descripción: Dada una tupla y un valor objetivo, encuentra el 
índice del primer elemento que coincida con el valor objetivo.'''

tupla_mix = (2, 7, 'Maria', True, 48, 'Irene', 'años', 'meses', False, 35)
tupla_mix.index("Maria")
```

```python
# Ejercicio 6: Contar ocurrencias de un elemento en una tupla

''' Descripción: Dada una tupla y un valor objetivo, cuenta cuántas 
veces aparece el valor objetivo en la tupla.'''

tupla_mix = (2, 7, 'Maria', True, 48, 'Irene', 'años', 'meses', False,"meses", 35)
print(tupla_mix.count(True))
print(tupla_mix.count("meses"))
```

```python
# Ejercicio 7: Obtener una subtuple de una tupla

''' Descripción: Dada una tupla, crea una nueva tupla que contenga los elementos 
desde el índice 1 hasta el índice 3 (sin incluirlo).'''

tupla_mix = (2, 7, 'Maria', True, 48, 'Irene', 'años', 'meses', False,"meses", 35)
subtupla = tupla_mix[1:3]
print(subtupla)
```

```python
# Ejercicio 8: Invertir una tupla

''' Descripción: Dada una tupla, inviértela en orden.'''

tupla_mix = (2, 7, 'Maria', True, 48, 'Irene', 'años', 'meses', False,"meses", 35)
tupla_inversa = list(tupla_mix)
tupla_inversa.reverse()
tupla_mix= tuple(tupla_inversa)
print(tupla_mix)

# no podemos invertir una tupla directamente, porque una tupla no se puede modificar, para poder hacerlo necesitariamos
# hacer una variable que tenga los mismos valores en una lista, aplicar sobre esta el metodo de listas reverse y 
# transformarlo en una tupla de nuevo bajo el mismo nombre. 
```
